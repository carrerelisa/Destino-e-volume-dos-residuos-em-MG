```python
# Devemos usar ponto ao invés de vírgula na casa decimal e o inverso nos milhares

df = pd.read_csv('/content/drive/MyDrive/Trabalho de Python/base.csv',
                 encoding='latin-1',
                 sep = ';',
                 decimal = ',',
                 thousands= '.')
```

```python
df.head() # Verificando se a importação está ok
```

```python
df.info() # Verificando se alterou o tipo de variável
# Variáveis de quantidade devem estar como float (ok)
```

```python
# Filtrando as colunas de interesse

df1 = df [ ['Município','Ano de Referência','UP003 - Tipo de unidade',
            'UP007 - Quantidade de RDO e RPU recebida na unidade de processamento' ] ].copy()
df1.copy()
```

```python
# Excluindo os missing values

df2= df1.dropna()
df2.copy()
```

```python
df2[df1['UP003 - Tipo de unidade']=='Lixão'] # Verificar tipo de unidade
```

```python
df2[df2['UP003 - Tipo de unidade'].str.contains('Lixão')].copy()
```

```python
# Filtrando por tipo de unidade - Lixão
df2_lixão=df2[df2['UP003 - Tipo de unidade'].str.contains('Lixão')].copy()
df2_lixão.copy()
```

```python
df3_lixão =df2_lixão.pivot_table(index=['Ano de Referência', 'Município' ],
                                 values =['UP007 - Quantidade de RDO e RPU recebida na unidade de processamento'],
                                 aggfunc='sum') # Pivotando a tabela ref a Lixão para depois agregar
```

```python
df3_lixão # Verificando o data frame pivotado
```

```python
# Agrupar todos os municípios para fazer análise do estado de Minas Gerais por tipo de unidade (Lixão)
df3_lixão = df3_lixão.groupby('Ano de Referência')['UP007 - Quantidade de RDO e RPU recebida na unidade de processamento'].sum()
```

```python
print(df3_lixão) # Verificar se o agrupamento para Lixão deu certo (ok)
```

```python
# Verificar nome das variáveis
df2['UP003 - Tipo de unidade'].unique()
```

```python
# Filtrando por tipo de unidade - Aterro Sanitário
df2[df2['UP003 - Tipo de unidade'].str.contains('sanitário')].copy()
df3_aterro=df2[df2['UP003 - Tipo de unidade'].str.contains('sanitário')].copy()
df3_aterro
```

```python
df3_aterro_sanitário =df3_aterro.pivot_table(index=['Ano de Referência', 'Município' ],
                                 values =['UP007 - Quantidade de RDO e RPU recebida na unidade de processamento'],
                                 aggfunc='sum') # Pivotando a tabela ref a Aterro Sanitário para depois agregar
```

```python
df3_aterro_sanitário
```

```python
# Agregando todos os municípos por ano, de acordo com tipo de unidade (Aterro Sanitário)

df3_aterrosanitário_mg = df3_aterro_sanitário.groupby('Ano de Referência')['UP007 - Quantidade de RDO e RPU recebida na unidade de processamento'].sum()
```

```python
df3_aterrosanitário_mg # Verificar se o agrupamento para Aterro Sanitário deu certo (ok)
```
